>I will declare all the list I want to be able to create the datamart. I need a list with songs ID to get the audio features

```python
song_id=[]
#song_duration=[]
song_name=[]
song_popularity=[]
song_danceability=[]
song_energy=[]
song_key=[]
song_loudness=[]
song_mode=[]
song_speechiness=[]
song_acousticness=[]
song_instrumentalness=[]
song_liveness=[]
song_valence=[]
song_tempo=[]
song_type=[]
song_id=[]
song_duration_ms=[]

#loop to get features and info for songs. 
for song in songs:
    song_id.append(song['track']['id']) #Track id
    
#I do it this way to avoid calling the function many times. one per song     
songs_details=sp.audio_features(tracks=song_id)
for song_det in songs_details:
    song_danceability.append(song_det['danceability'])
```
