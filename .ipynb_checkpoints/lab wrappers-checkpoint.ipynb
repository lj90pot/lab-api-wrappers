>I will declare all the list I want to be able to create the datamart. I need a list with songs ID to get the audio features

```python
# üè≠ Create a funtion to download the information I need 
def tracks_features(playlist_id):
    #Clean all variables
    song_id=[]
    #song_duration=[]
    song_name=[]
    song_popularity=[]
    song_danceability=[]
    song_energy=[]
    song_key=[]
    song_loudness=[]
    song_mode=[]
    song_speechiness=[]
    song_acousticness=[]
    song_instrumentalness=[]
    song_liveness=[]
    song_valence=[]
    song_tempo=[]
    song_type=[]
    song_id=[]
    song_duration_ms=[]

    #We get the playlist elements
    results = sp.user_playlist_tracks("spotify", playlist_id)
    songs = results['items']

    while results['next']:
        results = sp.next(results)
        songs.extend(results['items'])
   
    #loop to get features and info for songs.
    print(len(songs))
    songs=songs[:30]
    print("after",len(songs))
    for song in songs:
        song_id.append(song['track']['id']) #Track id
        song_name.append(song['track']['name']) #Track name
        song_popularity.append(song['track']['popularity']) #Track popularity
    
    dtc_song_id={'song_id':song_id}
    df_song_id =pd.DataFrame.from_dict(dtc_song_id)
    dtc_song_name={'song_name':song_name}
    df_song_name =pd.DataFrame.from_dict(dtc_song_name)
    dtc_song_popularity={'song_popularity':song_popularity}
    df_song_popularity =pd.DataFrame.from_dict(dtc_song_popularity)
    
    #I do it this way to avoid calling the function many times. one per song  
    songs_details=sp.audio_features(tracks=song_id) #get track features
    
    #Create a dataframe with the song features
    df = pd.DataFrame(songs_details)
    
    #Append tracks basic information
    df = pd.concat([df,df_song_name,df_song_popularity],axis=1)
    df.reset_index(drop=True)
    
    #Add the songs to the existing dataframe
    #data=pd.concat([df,data,axis=1)
    #data.reset_index(drop=True)
    
    return df
```
